```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#Define df as a dictionary
df = {'X': [1.2, 2.5, 3.6],
       'y': [8.5, 9.2 , 10.1]}
```

```python
#Make a pandas dataframe 
df=pd.DataFrame(df)
df
```

```python
# Define X and y as numpy array
X = np.asanyarray(df['X'])
y=np.asanyarray(df['y'])
```

```python
# Selecte 2 float number by using numpy random for assuming as weight and bias
b , w = np.random.randn(2)
print(b)
print(w)
```

```python
# Define the number of iterations that we want using in gradient descent
iter = 3
```

```python
# Define Learning rate to determine the step size at each iteration of the gradient descent
lr_rate = 0.05
```

```python
# Make a loop to perform gradient descent to minimize the cost function of a linear regression to update weight and bias based on  learning rate and cost function
#To plot the quadratic error and absolute error, firstly calculate the errors at each iteration then store these errors in a list to plot 
quad_errs = []
abs_errs = []
n = float(len(df['X']))
for i in range(iter): 
    y_pred = (w *df['X']) + b  
    er_w = (-2/n) * sum(df['X'] * (df['y'] - y_pred))
    er_b = (-2/n) * sum(df['y'] - y_pred) 
    w -= lr_rate * er_w  # Update w
    b -= lr_rate * er_b  # Update b
    
    quad_err = np.mean((df['y'] - (w * df['X'] + b)) ** 2)
    abs_err = np.mean(np.abs(df['y'] - (w * df['X'] + b)))
    quad_errs.append(quad_err)
    abs_errs.append(abs_err)
    
    print(f"w={w}, b={b}, quadratic error={quad_err}, absolute error={abs_err}")

plt.plot(range(1, iter+1), quad_errs, label="Quadratic Error")
plt.plot(range(1, iter+1), abs_errs, label="Absolute Error")
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.legend()
plt.show()
```

```python
## calculate quadratic error and absolute error for final weight and bias
mse = np.mean((df['y'] - y_pred)**2)
print('Mean Squared Error: ',mse)
mae = np.mean(np.abs(df['y'] - y_pred))
print('Mean Absolute Error: ',mae)
```
